# Score Range Analysis (Top-10 Shortlists)

This notebook quantifies how *spread out* similarity scores are **within each job’s top-10 shortlist**.  
For every job and model, we compute:

$$
\text{sim\_range} = \max(\text{similarity}) - \min(\text{similarity})
$$

Intuition: if ranges are consistently small, then the model treats the top-ranked candidates as near-equivalent, and the shortlist cutoff is likely amplifying very small score differences into hard inclusion/exclusion decisions.

```python
import glob
import os
import pandas as pd
import re
```

```python
BASE_DIR = "./similarity_matching_output"

models_dict = {
    "fair_lora": "Fair LoRA Adapter",
    "jb_v2": "JobBERT v2",
    "jb_v3": "JobBERT v3",
    "minilm_base": "MiniLM Job Matcher",
    "openai": "openAI Text embedding"
}
```

### Helper functions: locate model + compute job-level score ranges

This section defines two utilities:

- `extract_model_name(path)`: extracts the model folder name (e.g., `jb_v2`) from a file path.
- `process_file(path)`: loads a model’s `top10_all_jobs.csv`, aggregates rows by `job_id`, and writes a new file
  `top_sim_range_by_job.csv` containing min/max similarity and the resulting range per job.

```python
def extract_model_name(s):
    s = re.sub(r"/+", "/", s)
    s = re.sub(r"\\+", "/", s)
    s = s.rstrip("/")
    parts = [p for p in s.split("/") if p]
    return parts[-2]
    
def process_file(path: str) -> None:
    df = pd.read_csv(path)
    base_model = extract_model_name(path)

    if "job_id" not in df.columns:
        raise ValueError(f"{path}: Missing required column 'job_id'.")

    agg = (
        df.groupby("job_id", as_index=False)
          .agg(
              role=("role", "first"),
              sim_min=("similarity", "min"),
              sim_max=("similarity", "max"),
              n=("job_id", "size")
          )
    )
    agg["sim_range"] = agg["sim_max"] - agg["sim_min"]

    out_path = os.path.join(BASE_DIR, f"{base_model}/top_sim_range_by_job.csv")
    agg.to_csv(out_path, index=False)

    print(f"✓ {models_dict[base_model]} -> {out_path} ")
```

### Step 1: Generate `top_sim_range_by_job.csv` for each model

We search for all `top10_all_jobs.csv` files under `similarity_matching_output/*/`.  
For each model, we compute job-level similarity ranges (using the `process_file` function defined above) and save them back into the same model folder.

```python
pattern = os.path.join(BASE_DIR, "*/top10_all_jobs.csv")
files = sorted(glob.glob(pattern))

if not files:
    raise FileNotFoundError(f"No files matched: {pattern}")

for f in files:
    process_file(f)
```

### Step 2: Summarize ranges per model (job-level + role-level extremes)

Now that each model has a `top_sim_range_by_job.csv`, we compute a compact summary per model:

- **Mean** and **median** job-level range across all jobs
- **Smallest** and **largest** job-level range (with job IDs)
- Role-level extremes based on the **mean range per role**:
  - role with smallest mean range  
  - role with largest mean range  

Why both job-level and role-level?
- job-level extremes show the *most/least separable* shortlists
- role-level extremes reveal whether some roles systematically produce tighter or looser score clustering

```python
PATTERN = os.path.join(BASE_DIR, "*/top_sim_range_by_job.csv")
```

```python
def safe_idxmin(s: pd.Series):
    s2 = pd.to_numeric(s, errors="coerce")
    if s2.notna().any():
        return s2.idxmin()
    return None

def safe_idxmax(s: pd.Series):
    s2 = pd.to_numeric(s, errors="coerce")
    if s2.notna().any():
        return s2.idxmax()
    return None

def summarize_file(path: str) -> None:
    df = pd.read_csv(path)

    required = {"job_id", "role", "sim_range"}
    missing = required - set(df.columns)
    if missing:
        raise ValueError(f"{path}: missing columns {missing}. Found: {list(df.columns)}")

    df["sim_range"] = pd.to_numeric(df["sim_range"], errors="coerce")

    model = extract_model_name(path)
    
    mean_range = df["sim_range"].mean()
    median_range = df["sim_range"].median()

    # Smallest / largest by job_id (global)
    i_min = safe_idxmin(df["sim_range"])
    i_max = safe_idxmax(df["sim_range"])

    min_job_id = df.loc[i_min, "job_id"] if i_min is not None else None
    min_range = df.loc[i_min, "sim_range"] if i_min is not None else None

    max_job_id = df.loc[i_max, "job_id"] if i_max is not None else None
    max_range = df.loc[i_max, "sim_range"] if i_max is not None else None

    # Smallest / largest by role (based on role-level mean sim_range)
    role_stats = (
        df.groupby("role", as_index=False)["sim_range"]
          .mean()
          .rename(columns={"sim_range": "mean_sim_range"})
    )

    r_min = safe_idxmin(role_stats["mean_sim_range"])
    r_max = safe_idxmax(role_stats["mean_sim_range"])

    min_role = role_stats.loc[r_min, "role"] if r_min is not None else None
    min_role_range = role_stats.loc[r_min, "mean_sim_range"] if r_min is not None else None

    max_role = role_stats.loc[r_max, "role"] if r_max is not None else None
    max_role_range = role_stats.loc[r_max, "mean_sim_range"] if r_max is not None else None

    print(f"\n=== Model: {models_dict[model]} ===")
    print(f"Mean sim_range:   {mean_range:.6f}" if pd.notna(mean_range) else "Mean sim_range:   NA")
    print(f"Median sim_range: {median_range:.6f}" if pd.notna(median_range) else "Median sim_range: NA")

    print(f"Job_id with smallest range: {min_job_id}")
    print(f"Smallest range:             {min_range:.6f}" if pd.notna(min_range) else "Smallest range:             NA")

    print(f"Job_id with largest range:  {max_job_id}")
    print(f"Largest range:              {max_range:.6f}" if pd.notna(max_range) else "Largest range:              NA")

    print(f"Role with smallest range:   {min_role}")
    print(f"Smallest role mean range:   {min_role_range:.6f}" if pd.notna(min_role_range) else "Smallest role mean range:   NA")

    print(f"Role with largest range:    {max_role}")
    print(f"Largest role mean range:    {max_role_range:.6f}" if pd.notna(max_role_range) else "Largest role mean range:    NA")

    summary = {
        "model": models_dict[model],
        "mean_range": mean_range,
        "median_range": median_range,
        "min_job_id": min_job_id,
        "min_range": min_range,
        "max_job_id": max_job_id,
        "max_range": max_range,
        "min_role": min_role,
        "min_role_range": min_role_range,
        "max_role": max_role,
        "max_role_range": max_role_range
    }

    return summary
```

### Collect summaries into a single table

We store one summary record per model and combine them into a single DataFrame for easy inspection and plotting.

This table is the “source of truth” for the visualizations below.

```python
files = sorted(glob.glob(PATTERN))
if not files:
    raise FileNotFoundError(f"No output files matched: {PATTERN}")

models_summaries = []
for f in files:
    summary = summarize_file(f)
    models_summaries.append(summary)
```

```python
df = pd.DataFrame(models_summaries)
df
```

## Visualizing score range dispersion across models

The plots below provide different views of the same idea:  
**how tightly clustered similarity scores are inside the top-10 shortlist.**

```python
import matplotlib.pyplot as plt

bxp_stats = []

for _, row in df.iterrows():
    bxp_stats.append({
        "label": row["model"],
        "whislo": row["min_range"],        # lower whisker
        "q1": row["median_range"],         # collapsed box (no quartiles available)
        "med": row["median_range"],        # median
        "q3": row["median_range"],         # collapsed box
        "whishi": row["max_range"],        # upper whisker
        "fliers": []                       # no outliers
    })

fig, ax = plt.subplots(figsize=(10, 4))
ax.bxp(bxp_stats, showfliers=False)

ax.set_ylabel("sim_range")
ax.set_title("Within-shortlist similarity range by model (min–median–max)")
plt.xticks(rotation=0, ha="center")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plot_df = df.copy()

models = plot_df["model"].tolist()
x = np.arange(len(models))

# ---- Figure 1: Mean vs Median sim_range per model ----
means = plot_df["mean_range"].astype(float).values
medians = plot_df["median_range"].astype(float).values

plt.figure(figsize=(10, 4))
plt.scatter(x, means, label="Mean")
plt.scatter(x, medians, label="Median")
for i, (m, md) in enumerate(zip(means, medians)):
    plt.plot([i, i], [md, m])  # connect median<->mean

plt.xticks(x, models, rotation=0, ha="center")
plt.ylabel("sim_range")
plt.title("Within-shortlist similarity range (Mean vs Median) by model")
plt.legend()
plt.tight_layout()
plt.show()


# ---- Figure 2: Job-level extremes (min/max sim_range per model) ----
mins = plot_df["min_range"].astype(float).values
maxs = plot_df["max_range"].astype(float).values

plt.figure(figsize=(10, 4))
plt.scatter(x, mins, label="Min job range")
plt.scatter(x, maxs, label="Max job range")
for i, (mn, mx) in enumerate(zip(mins, maxs)):
    plt.plot([i, i], [mn, mx])  # min->max line
    # annotate the extreme job ids
    plt.annotate(str(plot_df.loc[i, "min_job_id"]), (i, mn), textcoords="offset points", xytext=(6, -10), ha="left")
    plt.annotate(str(plot_df.loc[i, "max_job_id"]), (i, mx), textcoords="offset points", xytext=(6, 6), ha="left")

plt.xticks(x, models, rotation=0, ha="center")
plt.ylabel("sim_range")
plt.title("Job-level similarity range extremes by model (min ↔ max)")
plt.legend()
plt.tight_layout()
plt.show()


# ---- Figure 3: Role-level extremes (min/max role mean sim_range per model) ----
role_mins = plot_df["min_role_range"].astype(float).values
role_maxs = plot_df["max_role_range"].astype(float).values

plt.figure(figsize=(10, 4))
plt.scatter(x, role_mins, label="Min role mean range")
plt.scatter(x, role_maxs, label="Max role mean range")
for i, (mn, mx) in enumerate(zip(role_mins, role_maxs)):
    plt.plot([i, i], [mn, mx])
    plt.annotate(str(plot_df.loc[i, "min_role"]), (i, mn), textcoords="offset points", xytext=(6, -10), ha="left")
    plt.annotate(str(plot_df.loc[i, "max_role"]), (i, mx), textcoords="offset points", xytext=(6, 6), ha="left")

plt.xticks(x, models, rotation=0, ha="center")
plt.ylabel("mean sim_range (by role)")
plt.title("Role-level similarity range extremes by model (min ↔ max)")
plt.legend()
plt.tight_layout()
plt.show()
```
